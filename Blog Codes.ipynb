```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_nse = pd.read_excel('NSE.xlsx')
data_nifty50 = pd.read_excel('NIFTY_50.xlsx')
data_bnifty = pd.read_excel('NIFTY_BANK.xlsx')
```

```python
df = pd.concat([data_nse, data_nifty50, data_bnifty], axis=0)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df = df.drop_duplicates(subset=['TIC'])
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.sort_values(['Sharpe Ratio', 'Annual Volatility'], ascending=[False, True], inplace=True)
```

```python
new_data = df[(df['Sharpe Ratio'] > 1.2) & (df['Annual Volatility'] < 0.52) & (df['MaxDD'] > -0.32) & (df['Latest Price'] <= 600)]
```

```python
new_data.to_excel('Nov_Invest.xlsx', index=False)
```
